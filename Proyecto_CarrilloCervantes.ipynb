### Para el segundo inciso "¿Qué sabor de helado serías?"

Después se creó la función "Helado", cuya función principal es hacerle al usuario 6 diferentes tipos de preguntas, para cada pregunta se creó una condicional la cual indica que si la respuesta es "tal", entonces una variable antes inicializada en cero, irá aumentando 1, pero si la respuesta es otra la variable se quedará igual. Una vez terminadas las preguntas, se creó otra condicional al final, la cual indica lo siguiente:

- Si la variable contador es igual a 6, entonces el sabor de helado será Vainilla.

- Si la variable contador es menor o igual a 5 y mayor o igual a 3, entonces el sabor de helado será Fresa.

- Pero si la variable contador en menor o igual a 2, entonces el sabor de helado será Chocolate.

```python
# Función Helado
# Te realiza 6 preguntas y con base al resultado te asigna que tipo de helado eres
def Helado():
	contador = 0
	print("\nResponde las siguientes preguntas:\n")
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
	respuesta = input('1.- ¿Estudias Ingeniería (y/n)? ')
	if respuesta == 'y' or respuesta == 'Y':
		contador += 1
	elif respuesta == 'n' or respuesta == 'N':
		contador = contador
	else:
		print('Esa respuesta no existe, por lo cual no se tomará en cuenta')
		contador = contador
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
	respuesta = input('2.- ¿Te consideras una persona feliz (y/n)? ')
	if respuesta == 'y' or respuesta == 'Y':
		contador += 1
	elif respuesta == 'n' or respuesta == 'N':
		contador = contador
	else:
		print('Esa respuesta no existe, por lo cual no se tomará en cuenta')
		contador = contador
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
	respuesta = input('3.- ¿Te gustan los tacos (y/n)? ')
	if respuesta == 'y' or respuesta == 'Y':
		contador += 1
	elif respuesta == 'n' or respuesta == 'N':
		contador = contador
	else:
		print('Esa respuesta no existe, por lo cual no se tomará en cuenta')
		contador = contador
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
	respuesta = input('4.- ¿Prefieres el pop o el rock? ')
	if respuesta == 'pop' or respuesta == 'Pop':
		contador += 1
	elif respuesta == 'rock' or respuesta == 'Rock':
		contador = contador
	else:
		print('Esa respuesta no existe, por lo cual no se tomará en cuenta')
		contador = contador
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
	respuesta = input('5.- ¿Qué prefieres perros o gatos? ')
	if respuesta == 'perros' or respuesta == 'Perros':
		contador += 1
	elif respuesta == 'gatos' or respuesta == 'Gatos':
		contador = contador
	else:
		print('Esa respuesta no existe, por lo cual no se tomará en cuenta')
		contador = contador
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
	respuesta = input('6.- ¿Prefieres los colores claros u obscuros? ')
	if respuesta == 'claros' or respuesta == 'Claros':
		contador += 1
	elif respuesta == 'obscuros' or respuesta == 'Obscuros':
		contador = contador
	else:
		print('Esa respuesta no existe, por lo cual no se tomará en cuenta')
		contador = contador
	# _ _ _ _ _ _ _ _ _ _ __ _ _ _ __ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _

	print('\n')

	if contador == 6:
		print('Tu sabor de helado es Vainilla')
	elif contador <=5 and contador >=3:
		print('Tu sabor de helado es Fresa')
	elif contador <=2:
		print('Tu sabor de helado es Chocolate')
```

### Para el tercer inciso "¿Cómo sería tu vida en Hogwarts?"

Se realizó una clase llamada "clase", la cual tiene como atributos varias características de una casa de Hogwarts. Con base a esta clase instanciamos las 4 casas de Hogwarts según sus características.

```python
# Clase "casa"
class casa:
	def __init__(self, fundador, cualidades, compañeros, mesa):
		self.fundador = fundador
		self.cualidades = cualidades
		self.compañeros = compañeros
		self.mesa = mesa

# Instanciamos
Gryffindor = casa("Godric Gryffindor", "Osadía, valentía y lealtad", "Harry Potter, Ron Weasley, Hermione Granger", 3)
Hufflepuff = casa("Helga Hufflepuff", "Honestidad, trabajo duro y paciencia", "Cedric Diggory, Nymphadora Tonks", 4)
Ravenclaw = casa("Rowena Ravenclaw", "Originalidad, ingenio e inteligencia", "Luna Lovegood, Cho Chang", 2)
Slytherin = casa("Salazar Slytherin", "Ambición, astucia y determinación", "Voldemort, Draco Malfoy", 1)
```

```python
def Hogwarts(nombre):
	print('\n\t**********************************************************')
	print('\t\tHola, '+nombre)
	print('\tPara saber como sería tu vida en Hogwarts debes de')
	print('\tresponder 1 sencilla pregunta')
	print('\t _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ')
	print('\t|Estas en un examen de posiciones con el profesor Snape |')
	print('\t|pero no recuerdas ninguna respuesta, ¿Qué haces?       |')
	print('\t|_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _|')
	respuesta = input("""\t  a) Trato de recordar mis apuntes
\t  b) Nunca pasaría eso, por que si estudie
\t  c) Le pediría ayuda al profesor aun que me aterra preguntarle
\t  d) Le diría a un amigo que me pase la respuesta\n\n  Respuesta: """)

	if respuesta == 'a' or respuesta == "A":
		print('\n    | Genial '+nombre+', tu casa es Hufflepuff')
		print('    | El fundador de tu casa es: ', Hufflepuff.fundador)
		print('    | Las principales cualidades de los miembros de tu casa son: ', Hufflepuff.cualidades)
		print('    | Tus compañeros son: ', Hufflepuff.compañeros)
		print('    | Tu casa tiene la mesa número ', Hufflepuff.mesa, ' en el gran comedor\n')
	elif respuesta == 'b' or respuesta == "B":
		print('\n    | Genial '+nombre+', tu casa es Ravenclaw')
		print('    | El fundador de tu casa es: ', Ravenclaw.fundador)
		print('    | Las principales cualidades de los miembros de tu casa son: ', Ravenclaw.cualidades)
		print('    | Tus compañeros son: ', Ravenclaw.compañeros)
		print('    | Tu casa tiene la mesa número ', Ravenclaw.mesa, ' en el gran comedor\n')
	elif respuesta == 'c' or respuesta == "C":
		print('\n    | Genial '+nombre+', tu casa es Gryffindor')
		print('    | El fundador de tu casa es: ', Gryffindor.fundador)
		print('    | Las principales cualidades de los miembros de tu casa son: ', Gryffindor.cualidades)
		print('    | Tus compañeros son: ', Gryffindor.compañeros)
		print('    | Tu casa tiene la mesa número ', Gryffindor.mesa, ' en el gran comedor\n')
	elif respuesta == 'd' or respuesta == "D":
		print('\n    | Genial '+nombre+', tu casa es Slytherin')
		print('    | El fundador de tu casa es: ', Slytherin.fundador)
		print('    | Las principales cualidades de los miembros de tu casa son: ', Slytherin.cualidades)
		print('    | Tus compañeros son: ', Slytherin.compañeros)
		print('    | Tu casa tiene la mesa número ', Slytherin.mesa, ' en el gran comedor\n')
	else:
		print('\n\tEsa respuesta no existe ):')
```

## Menú de opciones

Finalmente, se creó un ciclo While el cual se irá repitiendo un menú de juegos hasta que el usuario decida salir. Para cada opción se llamo a la función correspondiente previamente comentadas; sin embargo, para la opción "salir" se creó una condicionante anidada, en la cual el usuario debe de verificar 2 veces que quiere salir del programa. La salida del programa se encuentra posteriormente.

```python
# Ciclo while, se va a repetir hasta que el usuario elija la opción salir
while True:
	try:
		print('\n________________________ JUEGUITOS ________________________')
		print('\n¿Qué quieres hacer?\n  1) Predicciones según tu horoscopo\n  2) Saber que tipo de helado serías\n  3) ¿Cómo sería tu vida en Hogwarts?\n  4) Salir')
		op = int(input("Opción: "))

		if op == 1:
			print('_ _ _ _ _ _ _ _ _ _ _ _ \n')
			print('\tPredicciones...\n')
			print(DHoroscopo.keys())
			signo = str(input("\n¿Cuál es tu signo Zodiacal? "))
			Horoscopo(signo)
		elif op == 2:
			print('_ _ _ _ _ _ _ _ _ _ _ _ ')
			print('\n¿Qué tipo de helado eres?')
			Helado()
		elif op == 3:
			print('_ _ _ _ _ _ _ _ _ _ _ _ ')
			print('\n¿Cómo sería tu vida en Hogwarts?')
			nombre = str(input('Escribe tu nombre: '))
			Hogwarts(nombre)
		elif op == 4:
			salir = input('\n¿Enserio quieres salir (y/n)? ')
			if salir == 'n' or salir == 'N':
				print('\nSabía que no te querias ir (:')
			elif salir == 'y' or salir == 'Y':
				salir = input('\n¿Neta me lo juras (y/n)? ')
				if salir == 'n' or salir == 'N':
					print('\nSabía que no te querias ir (:')
				elif salir == 'y' or salir == 'Y':
					print('\nOk, byeeee ):')
					break
				else:
					print('\nEsa opción no existe, entonces creo no te quieres ir (:')
			else:
				print('\nEsa opción no existe, entonces creo no te quieres ir (:')
		else:
			print('\nNo existe esa opción ):')
		
	except ValueError:
		print('\nNo existe esa opción ):')
```
